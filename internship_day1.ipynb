```python
df['age_mean_inputed']=df['Age'].fillna(df['Age'].mean())
df['age_median_inputed']=df['Age'].fillna(df['Age'].median())
```

```python
df['cabin_mode_inputted']=df['Cabin'].fillna(df['Cabin'].mode()[0])
```

creating box plot

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Embarked',y='Fare',palette='pastel')

plt.title('Boxplot of Age by Passenger Class',fontsize=12)
plt.xlabel('Passenger Class',fontsize=12)
plt.ylabel('Survived',fontsize=12)
plt.show()
```

```python
encoded_data=df.copy()
encoded_data
```

label encoding for sex

```python
label_encoder=LabelEncoder()
encoded_data['Sex_Label']=label_encoder.fit_transform(encoded_data['Sex'])
encoded_data['Embarked']=label_encoder.fit_transform(encoded_data['Embarked'])
encoded_data['Sex_Label']
```

one hot encoding for embarked

```python
one_hot_encoded_data=pd.get_dummies(encoded_data['Embarked'],prefix='Embarked')
encoded_data=pd.concat([encoded_data,one_hot_encoded_data],axis=1)
encoded_data
```
