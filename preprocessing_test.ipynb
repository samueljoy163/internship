```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler,StandardScaler

```

```python
df=pd.read_csv("/content/preprocessing_practice_dataset.csv")
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Age_mean_inputed']=df['Age'].fillna(df['Age'].mean())
df['month_salary_mean_inputed']=df['Monthly_Salary'].fillna(df['Monthly_Salary'].mean())
```

```python
df['Department_m']=df['Department'].fillna(df['Department'].mode()[0])
df['Education_m']=df['Education'].fillna(df['Education'].mode()[0])
df['city_m']=df['City'].fillna(df['City'].mode()[0])
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(30,20))
sns.boxplot(data=df,x='Employee_ID',y='Department',palette='pastel')

plt.title('Boxplot of Age by experience',fontsize=20)
plt.xlabel('employee',fontsize=20)
plt.ylabel('department name',fontsize=20)
plt.show()
```

```python

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Department_m', palette='viridis')
plt.title('Number of Employees per Department', fontsize=16)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(x=df['Age'].dropna(), bins=15)
plt.title('Histogram of Age Distribution', fontsize=12)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x=df['Experience_Years'], y=df['Monthly_Salary'], color='royalblue')
plt.title('Monthly Salary vs. Experience Years')
plt.xlabel('Experience Years')
plt.ylabel('Monthly Salary')
plt.show()
```

```python

sns.set_theme(style="whitegrid")
dept_counts = df['Department'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(x=dept_counts.values,
        labels=dept_counts.index)
plt.title('Pie Graph of Department Distribution', fontsize=14)
plt.show()
```
